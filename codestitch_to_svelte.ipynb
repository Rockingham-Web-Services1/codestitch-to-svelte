```python
import copy
name = "banner_559"


# Load the HTML file
with open(html_filename, 'r') as file:
    html_content = file.read()

# Create a BeautifulSoup object
soup = BeautifulSoup(html_content, 'html.parser')

# Content derived from the original html_content
variables_list = []

"""
Title <span> Tag
"""

# Find all spans which include "title" or "topper" + modify
all_span = soup.find_all('span')

# Iterate through each <span> tag
for span in all_span:
    str_span = str(span)
    # Check if the 'class' attribute exists and retrieve it
    if 'class' in span.attrs:
        # Check if the class attribute contains 'topper' or 'title'
        if 'topper' in str_span:
            # Change the text inside the <span> tag
            span.string = "{subtitle}"
            variables_list.append("{subtitle}")
        elif 'title' in str_span:
            span.string = "{title}"
            variables_list.append("{title}")
        else: 
            pass

"""
Links <a> Tag
"""

# Find all links
all_a = soup.find_all('a')

# Iterate through each <a> tag
i = 0
for a in all_a:
    link_var = "{link" + str(i) + "}"
    variables_list.append(link_var)
    a['href'] = a['href'].replace("", link_var)
    i += 1
    
"""
Links <img> <source> Tag
"""

# Find all img tags
all_img = soup.find_all('img')

# Iterate through each <source> tag
i = 0
for img in all_img:
    str_img = str(img)
    src_var = "{image" + str(i) + "}"
    img['src'] = src_var
    img['height'] = "{height}"
    img['width'] = "{width}"
    img['alt'] = "{alt}"
    variables_list.append(src_var)
    variables_list.append("{height}")
    variables_list.append("{width}")
    variables_list.append("{alt}")
    i += 1

# Find all source tags
all_source = soup.find_all('source')

# Iterate through each <source> tag
i = 0
for source in all_source:
    str_source = str(source)
    srcset_var = "{source" + str(i) + "}"
    variables_list.append(srcset_var)
    source['srcset'] = srcset_var
    i += 1

"""
Hard coded text
"""
# The content to be written to the svelte file
variable_string = ""
i = 0
for s in variables_list:
    start = "{"
    end = "}"
    new_s = (s.split(start))[1].split(end)[0]
    if i == 0:
        variable_string += new_s
    else: 
        variable_string = f"{variable_string}, {new_s}"
    i += 1
script_tag = f"<script>\n    export let {variable_string};\n</script>"
stitch_name = "Banner 559"
comment = f"<!-- ============================================ -->\n<!--                {stitch_name}                  -->\n<!-- ============================================ -->"

"""
Final Content to put in svelte file
"""
# Once all modifications are done, convert the soup object back to string
modified_html_content = str(soup)
svelte_file_content = f"""{script_tag}\n\n{comment}\n\n{modified_html_content}"""


"""
-------------------------------------------------------------
JSON FILES
-------------------------------------------------------------
"""

"""
Write variables into python variables
"""

for var in variables_list:
    start = "{"
    end = "}"
    new_var = (var.split(start))[1].split(end)[0]
    print(f"{new_var} = \"\"")


subtitle = ""
title = ""
link0 = ""
link1 = ""
image0 = ""
height = ""
width = ""
alt = ""
source0 = ""
source1 = ""

"""
JSON file --> Needed to make the alt, height and width children elements
"""

JSON_str = ""

"""
Write to files
"""

with open(svelte_filename, "w") as f:
    f.write(svelte_file_content)
    
```

```python
import os
from bs4 import BeautifulSoup
from itertools import chain

name = "contact_486"
html_filename = f"{name}/{name}.html"
css_filename = f"{name}/{name}.css"
svelte_filename = f"{name}_modified/banner_559.svelte"
json_filename = f"{name}_modified/{name}.json"

# Load the HTML file
with open(html_filename, 'r') as file:
    html_content = file.read()
    
# Create a BeautifulSoup object
soup = BeautifulSoup(html_content, 'html.parser')

var_list = []

"""
<span> tag 
- title 
- topper
- color
- heading
"""

# Find all spans 
all_span = soup.find_all('span')

# Iterate through each <span> tag
subtitle_i = 0
title_i = 0
color_i = 0
heading_i = 0

for span in all_span:
    str_span = str(span)
    # Check if the 'class' attribute exists and retrieve it
    if 'class' in span.attrs:
        # Check if the class attribute contains 'topper' or 'title'
        if 'topper' in str_span:
            # Change the text inside the <span> tag
            subtitle_var = "{subtitle" + str(subtitle_i) + "}"
            span.string = subtitle_var
            variables_list.append([subtitle_var])
            subtitle_i += 1
        elif 'title' in str_span:
            title_var = "{title" + str(title_i) + "}"
            span.string = title_var
            variables_list.append([title_var])
            title_i += 1
        elif 'color' in str_span:
            color_var = "{color" + str(color_i) + "}"
            span.string = color_var
            variables_list.append([color_var])
            color_i += 1
        elif 'heading' in str_span:
            heading_var = "{heading" + str(heading_i) + "}"
            span.string = heading_var
            variables_list.append([heading_var])
            heading_i += 1
        else: 
            pass

"""
<h2> tag 
- title 
"""
# Find all <h2> tags
all_h2 = soup.find_all('h2')
i = 0
for h2 in all_h2:
    str_h2 = str(h2)
    h2_var = "{h2_" + str(i) + "}"
    variables_list.append([h2_var])
    h2.string = h2_var
    i += 1

"""
<h3> tag 
"""

# Find all <h3> tags
all_h3 = soup.find_all('h3')
i = 0
for h3 in all_h3:
    str_h3 = str(h3)
    h3_var = "{h3_" + str(i) + "}"
    variables_list.append([h3_var])
    h3.string = h3_var
    i += 1

"""
<p> tag 
- text 
"""

# Find all <p> tags
all_p = soup.find_all('p')
i = 0
for p in all_p:
    str_p = str(p)
    p_var = "{p_" + str(i) + "}"
    variables_list.append([p_var])
    p.string = p_var
    i += 1
    
"""
<a> tag 
- href
- may or may not have a cs-link class
"""

# Find all links
all_a = soup.find_all('a')

# Iterate through each <a> tag
i = 0
for a in all_a:
    href_var = "{link" + str(i) + "}"
    a_var = "{link_text" + str(i) + "}"
    variables_list.append([href_var, a_var])
    a['href'] = link_var
    a.string = a_var
    i += 1

"""
<input> tag 
May have the following attributes
- name
- placeholder
"""

# Find all inputs
all_input = soup.find_all('input')
input_name_var_i = 0
input_placeholder_var_i = 0
input_type_var_i = 0

for _input in all_input:
    input_name_var = "{input_name" + str(input_name_var_i) + "}"
    input_placeholder_var = "{input_placeholder" + str(input_placeholder_var_i) + "}"
    input_type_var = "{input_placeholder" + str(input_type_var_i) + "}"
    variables_list.append([input_name_var, input_placeholder_var, input_type_var])
    _input['name'] = input_name_var
    _input['placeholder'] = input_placeholder_var
    _input['type'] = input_type_var
    input_name_var_i += 0
    input_placeholder_var_i += 0
    input_type_var_i += 0

"""
<option> tag 
- value
"""

# Find all <option> tags
all_option = soup.find_all('option')

# Iterate through each <option> tag
i = 0
option_value_i = 0
for option in all_option:
    option_var = "{option" + str(i) + "}"
    option_value_var = "{option_value" + str(option_value_i) + "}"
    variables_list.append([option_var, option_value_var])
    option.string = option_var
    option['value'] = option_value_var
    i += 1
    option_value_i += 1

"""
<button> tag 
- value
"""
# Find all <button> tags
all_button = soup.find_all('button')
i = 0
button_type_var_i = 0
for button in all_button:
    button_var = "{button" + str(i) + "}"
    button_type_var = "{button_type" + str(button_type_var_i) + "}"
    variables_list.append([button_var, button_type_var])
    button['type'] = button_type_var
    a.string = button_var
    i += 1
    button_type_var_i += 1

"""
<label> tag --> it's replacing the input tag as well as the text 
"""

'''
# Find all links
all_label= soup.find_all('label')
i = 0
for label in all_label:
    label_var = "{label" + str(i) + "}"
    variables_list.append(label_var)
    label.contents[0] = label_var
    i += 1

print(all_label[0])
'''

# The content to be written to the svelte file
variable_string = ""
i = 0
# converting 2d list into 1d
# using chain.from_iterables
flatten_list = list(chain.from_iterable(variables_list))
print(flatten_list)
for s in flatten_list:
    start = "{"
    end = "}"
    new_s = (s.split(start))[1].split(end)[0]
    if i == 0:
        variable_string += new_s
    else: 
        variable_string = f"{variable_string}, {new_s}"
    i += 1

script_tag = f"<script>\n    export let {variable_string};\n</script>"
comment = f"<!-- ============================================ -->\n<!--                {name}                  -->\n<!-- ============================================ -->"

"""
Final Content to put in svelte file
"""
# Once all modifications are done, convert the soup object back to string
modified_html_content = str(soup)
svelte_file_content = f"""{script_tag}\n\n{comment}\n\n{modified_html_content}"""

with open(svelte_filename, "w") as f:
    f.write(svelte_file_content)
    
```
