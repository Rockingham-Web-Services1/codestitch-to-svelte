```python
import os
from bs4 import BeautifulSoup
from collections import Iterable

def flatten(lis):
    for item in lis:
        if isinstance(item, Iterable) and not isinstance(item, str):
            for x in flatten(item):
                yield x
        else:        
            yield item


stitch_name = "contact_1392"
html_filename = f"{stitch_name}/{stitch_name}.html"
svelte_filename = f"{stitch_name}/{stitch_name}_modified/{stitch_name}.svelte"

# Load the HTML file
with open(html_filename, 'r') as file:
    html_content = file.read()
    
# Create a BeautifulSoup object
soup = BeautifulSoup(html_content, 'html.parser')

var_list = []

"""
Modify tags
"""
class HTMLModifier: 
    def __init__(self, soup):
        self.soup = soup
        self.variables_list = []
    
    def get_modified_html(self, stitch_name):
        variable_str = ""
        flatten_var_list = flatten(self.variables_list)
        print(flatten_var_list)
        i = 0
        for s in flatten_var_list:
            start = "{"
            end = "}"
            new_s = (s.split(start))[1].split(end)[0]
            if i == 0:
                variable_str += new_s
            else: 
                variable_str = f"{variable_str}, {new_s}"
            i += 1
        
        script_tag = f"<script>\n    export let {variable_str};\n</script>"
        comment = f"<!-- ============================================ -->\n<!--                {stitch_name}                  -->\n<!-- ============================================ -->"
        modified_html_content = str(soup)
        svelte_file_content = f"""{script_tag}\n\n{comment}\n\n{modified_html_content}"""
        return svelte_file_content
    
    # span tags 
    def modify_span_tags(self, soup):
        # Find all spans 
        all_span = soup.find_all('span')
        
        subtitle_i = 0
        title_i = 0
        color_i = 0
        heading_i = 0
        header_i = 0
        
        for span in all_span: 
            str_span = str(span)
            if 'topper' in str_span:
                self.modify_span_class_content(span, json_var_name="subtitle", i=subtitle_i)
                subtitle_i += 1
            elif 'title' in str_span:
                self.modify_span_class_content(span, json_var_name="title", i=title_i)
                title_i += 1
            elif 'color' in str_span:  
                self.modify_span_class_content(span, json_var_name="color", i=color_i)
                color_i += 1
            elif 'heading' in str_span:  
                self.modify_span_class_content(span, json_var_name="heading", i=heading_i)
                heading_i += 1
            elif 'header' in str_span:  
                self.modify_span_class_content(span, json_var_name="header", i=header_i)
                header_i += 1
            else:
                pass
    def modify_span_class_content(self, span, json_var_name, i):
        json__var = "{" + str(json_var_name) + str(i) + "}"
        span.string = json__var
        self.variables_list.append(json__var)
        
    # p tags    
    def modify_p_tags(self, soup):
        all_tag = soup.find_all('p')
        i = 0
        for tag in all_tag:
            json_var_name="p"
            self.modify_p_content(tag, json_var_name, i)
            i += 1   
    def modify_p_content(self, tag, json_var_name, i):
        json__var = "{" + str(json_var_name) + "_"+ str(i) + "}"
        tag.string = json__var
        self.variables_list.append(json__var)
        
    # h1 tags
    def modify_h1_tags(self, soup):
        all_tag = soup.find_all('h1')
        i = 0
        for tag in all_tag:
            json_var_name="h1"
            self.modify_p_content(tag, json_var_name, i)
            i += 1   
    
    # h2 tags
    def modify_h2_tags(self, soup):
        all_tag = soup.find_all('h2')
        i = 0
        for tag in all_tag:
            json_var_name="h2"
            self.modify_p_content(tag, json_var_name, i)
            i += 1   
            
    # h3 tags
    def modify_h3_tags(self, soup):
        all_tag = soup.find_all('h3')
        i = 0
        for tag in all_tag:
            json_var_name="h3"
            self.modify_p_content(tag, json_var_name, i)
            i += 1   
    
    # a tags
    def modify_a_tags(self,soup):
        all_tag = soup.find_all('a')
        i = 0
        
        for tag in all_tag:
            self.modify_a_content(tag, json_var_str_1 = "link_tree", json_var_1 = "url", i=i)          
            i += 1      
    def modify_a_content(self, tag, json_var_str_1, json_var_1, i):
        json_var_str_1 = "{" + str(json_var_str_1) + "_"+ str(i) + "}"
        json_var_1 = "{" + str(json_var_1) + "_"+ str(i) + "}"
        tag.string = json_var_str_1
        tag['href'] = json_var_1 
        self.variables_list.append([json_var_str_1, json_var_1])
    
    # input tags
    def modify_input_tags(self, soup):
        all_tag = soup.find_all('input')
        i =  0
        for tag in all_tag:
            self.modify_input_content(tag, json_var_1="input_name", json_var_2="placeholder",json_var_3="type", i=i)          
            i += 1   

    def modify_input_content(self, tag, json_var_1, json_var_2,json_var_3, i):
        json_var_1 = "{" + str(json_var_1) + "_"+ str(i) + "}"
        json_var_2 = "{" + str(json_var_2) + "_"+ str(i) + "}"
        json_var_3 = "{" + str(json_var_3) + "_"+ str(i) + "}"
        tag["name"] = json_var_1 
        tag["placeholder"] = json_var_2 
        tag["type"] = json_var_3
        self.variables_list.append([json_var_1, json_var_2, json_var_3])
    
    # Option tags
    def modify_option_tags(self, soup):
        all_tag = soup.find_all('option')
        i =  0
        for tag in all_tag:
            self.modify_optio_content(tag, json_var_str_1 = "option_text", json_var_1 = "option_value", i=i)             
            i += 1
            
    def modify_option_content(self, tag, json_var_str_1, json_var_1, i):
        json_var_str_1 = "{" + str(json_var_str_1) + "_"+ str(i) + "}"
        json_var_1 = "{" + str(json_var_1) + "_"+ str(i) + "}"
        tag.string = json_var_str_1
        tag['value'] = json_var_1 
        self.variables_list.append([json_var_str_1, json_var_1])

    
    # Option tags
    def modify_button_tags(self, soup):
        all_tag = soup.find_all('button')
        i =  0
        for tag in all_tag:
            self.modify_option_content(tag, json_var_str_1 = "button", json_var_1 = "button_type", i=i)             
            i += 1
            
    def modify_button_content(self, tag, json_var_str_1, json_var_1, i):
        json_var_str_1 = "{" + str(json_var_str_1) + "_"+ str(i) + "}"
        json_var_1 = "{" + str(json_var_1) + "_"+ str(i) + "}"
        tag.string = json_var_str_1
        tag['button'] = json_var_1 
        self.variables_list.append([json_var_str_1, json_var_1])
        
    # image tag
    # label tag
    # picture tag
"""
Final Content to put in svelte file
"""
modifier = HTMLModifier(soup)
modifier.modify_span_tags(soup)
modifier.modify_p_tags(soup)
modifier.modify_h2_tags(soup)
modifier.modify_h3_tags(soup)
modifier.modify_a_tags(soup)
modifier.modify_input_tags(soup) 
modifier.modify_option_tags(soup) 
modifier.modify_button_tags(soup)
svelte_file_content = modifier.get_modified_html(stitch_name)

with open(svelte_filename, "w") as f:
    f.write(svelte_file_content)
```
